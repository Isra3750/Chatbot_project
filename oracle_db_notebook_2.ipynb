```python
# Querying by vector distance
sql = """
    SELECT movie_quote, movie, movie_type, movie_year, distance
    FROM (
         SELECT movie_quote, movie, movie_type, movie_year, 
                vector_distance(
                    movie_quote_vector,
                    vector_embedding(all_minilm_l12_v2 USING :search_text as data)
                ) AS distance
         FROM movie_quotes
         ORDER BY distance
    ) t
    WHERE ROWNUM <= 3
"""
conn = oracledb.connect(user="testuser1", password="testuser1", dsn="localhost:1521/freepdb1")
cur = conn.cursor()
cur.execute(sql, search_text="Sad quotes")
rows = cur.fetchall()
for row in rows:  # Use 'rows' instead of 'cursor'
    print(row)
cur.close()
conn.close()
```
